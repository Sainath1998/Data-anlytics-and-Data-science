```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#data_import = pd.read_csv()
```

```python
athletes = pd.read_csv("athlete_events.csv")
region = pd.read_csv("noc_regions.csv")
```

```python
athletes.head() #Displayes top 5 elements of the table athletes
```

```python
region.head() #Displayes top 5 elements of the table Region
```

```python
athletes_df = athletes.merge(region , how = 'left', on = 'NOC') #merge two datasets horizontally
athletes_df.head()
```

```python
athletes_df.shape  #displays new no, of rows and column
```

```python
#column names consistent
athletes_df.rename(columns={'region':'Region','notes':'Notes'},inplace = True)
```

```python
# top countries participating in olympics
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries.index
```

```python
# plot bargraph
plt.figure(figsize=(12,6))
#plt.xticks(rotation = 20)
plt.title('Overall participation by the countries')
sns.barplot(x=top_10_countries.index,y=top_10_countries.values,palette = 'Set2');
```

```python
import plotly.express as px #importing plotly libraries
```

```python
px.histogram(athletes_df,x=top_10_countries.index, y = top_10_countries.values,color=top_10_countries) # histogram using plotly
```

```python
#age distribution
age_dist = athletes_df.Age.value_counts().sort_values(ascending=False).head(10)
age_dist
```

```python
px.histogram(athletes_df,x=age_dist.index,y=age_dist.values,color=age_dist.index)
```
